```python
import pandas as pd
```

```python
df = pd.read_csv('/content/titanic.csv')
df.head()
```

```python
#Dropoping irrelevant features
df.drop(['PassengerId','Name','Ticket','SibSp','Parch','Cabin','Embarked'],axis=1,inplace=True)
df.head()
```

```python
#seperating out the target variable
target = df['Survived']
inputs = df.drop('Survived',axis=1)
inputs
```

```python
#encoding male and female
inputs['Sex'] = inputs['Sex'].map({'male': 1, 'female': 0})
inputs.head()
```

```python
#checking for null values
inputs.columns[inputs.isna().any()]
```

```python
#Filling Null values with the mean of the age
inputs['Age'] = inputs['Age'].fillna(inputs['Age'].mean())
inputs.head()
```

```python

#splitting dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(inputs,target,test_size=0.2,random_state=32)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_test.head(10)
```

```python
model.predict(X_test)
```

```python
#checking models performance on test data
model.score(X_test,Y_test)
```

```python
#checking the probability stats
model.predict_proba(X_test.head(10))
```
