IMDB SENTIMENT ANALYSIS USING LOGISTIC REGRESSION


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Prade\OneDrive\Desktop\YOGI STUDY\NLP\IMDB Dataset.csv')
df
```

```python
def encode(n):
    if n=='positive':
        return 1
    else:
        return 0
df['sentiment'] = df['sentiment'].apply(encode)  
df 
```

```python
#hyperparamter tuning 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',
    token_pattern=r'[a-zA-Z]{2,}',
    max_features=20000)
```

```python
target = df['sentiment']
Features = df['review']
Features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(Features,target,test_size=0.2,random_state=32)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape

```

```python
#fitting train data and transforming test data
X_train = cv.fit_transform(X_train)
X_test = cv.transform(X_test)
```

```python
# performing sentiment analysis using Logistic Regression this time
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train,Y_train)
print("Model score :",model.score(X_test,Y_test))
```

```python
#fetching the predicted values by the model on test set
y_pred = model.predict(X_test)
```

```python
# checking the model performace using classification reports
#Precision = (TP)/TP + FP , recall = TP/TP+FN 
#precison denominator represents total postives predicted,how many were truly positive,How many Predicted positives are right
# recall denominator represents total actual positives,how many were truly positive,How many Real positives did we catch
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(y_pred, Y_test)
print(score)
print(classification_report(y_pred , Y_test ))
```

```python
#displaying confusion matarices
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
conf_matrix = confusion_matrix(y_pred, Y_test)
ConfusionMatrixDisplay(conf_matrix).plot()
```
