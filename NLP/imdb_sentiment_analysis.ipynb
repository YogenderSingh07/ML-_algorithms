```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Prade\OneDrive\Desktop\YOGI STUDY\NLP\IMDB Dataset.csv')
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(df['review'])
```

```python
# trying hyperparamter tuning 
cx = CountVectorizer(stop_words='english',
    token_pattern=r'[a-zA-Z]{2,}',
    max_features=20000)
```

```python
#gives vocab consisting of only english words 
V = cx.fit_transform(df['review'])
print(cx.get_feature_names_out())
```

```python
#checking the features without hyperparameter tuning,gives vocab with words from hebrew,polish and digits aswell
print(cv.get_feature_names_out())
```

```python
def encode(n):
    if n=='positive':
        return 1
    else:
        return 0
df['sentiment'] = df['sentiment'].apply(encode)  
df 
```

```python
target = df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(V,target,test_size=0.2,random_state=32)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
#further metrics to check model performance 
from sklearn.metrics import classification_report

print(classification_report(Y_test, y_pred))
```
